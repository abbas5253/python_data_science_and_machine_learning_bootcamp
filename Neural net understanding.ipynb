```python
IMG_SHAPE = (224, 224, 3)
model = tf.keras.applications.VGG16(weights='imagenet', include_top=False, input_shape=IMG_SHAPE)
```

```python
tf.reset_default_graph()
sess= tf.InteractiveSession()
```

```python
model.summary()
```

```python
int_set = sess.run(model.get_layer('block2_pool').output, feed_dict={model.input:img.reshape(1,224,224,3)})
```
